```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\HP\Downloads\chemical_plant_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Perform grouping and calculate mean
grouped_means = df.groupby('Temperature',)['Pressure',].mean()
print(grouped_means)
```

```python
# Fill a column with its mean (for normally distributed data)
df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mean())

# Fill a column with its median (for data with outliers)
df['Pressure'] = df['Pressure'].fillna(df['Pressure'].median())

# Fill a column with its mode (for categorical data)
df['Flow_Rate'] = df['Flow_Rate'].fillna(df['Flow_Rate'].mode()[0])
```

```python
import matplotlib.pyplot as plt

# Plotting temperature vs pressure
plt.plot(df['Temperature'], df['Pressure'])
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.title('Chemical_plant_dataset')
plt.show()
```

```python
# Plotting a bar chart for Temperature bby Pressure
df.groupby('Temperature')['Pressure'].mean().plot(kind='bar')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.title('Average Temperature by Pressure')
plt.show()
```

```python
# Plotting a histogram for FlowRate
df['Flow_Rate'].plot(kind='hist', bins=10)
plt.xlabel('Flow_Rate')
plt.title('Flow_Rate')
plt.show()
```

```python
# Plotting a scatter plot of Flowrate vs Hazard
df.plot(kind='scatter', x='Flow_Rate', y='Hazard')
plt.title('Flowrate vs hazard')
plt.show()
```
